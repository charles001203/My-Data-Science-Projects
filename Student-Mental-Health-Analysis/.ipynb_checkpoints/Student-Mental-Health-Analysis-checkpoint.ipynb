```python
# Merging values in Education Level
def categorizeEducationLevel(edu_lvl):
    if edu_lvl in ['MTech', 'MSc', 'MA']:
        return "Masters"
    elif edu_lvl in ['BSc', 'BTech', 'BA']:
        return "Bachelors"
    elif edu_lvl in ['Class 8']:
        return "Middle School"
    else:
        return "High School"        
```

```python
df['Education Level'] = df['Education Level'].map(categorizeEducationLevel)
df.head()
```

```python
print(df['Education Level'].value_counts())
```

#### Replacing the values in columns with numerical values
* Gender: Male = 1, Female = 0, Other = 2
* Education Level: Masters = 0, Bachelors = 1, High School = 2, Middle School = 3
* Stress Level: Low = 0, medium = 1, high = 2
* Anxious before exam: Yes = 1, No = 0
* Academic Performance Change: Same = 0, Improved = 1, Declined = 2

### 2.3 Data Preprocessing

#### Label Encoding the Categorical Variables

```python
le = LabelEncoder()
cols = ['Gender', 'Education Level', 'Stress Level', 'Anxious Before Exams','Academic Performance Change']
for i in cols:
    le.fit(df[i])
    df[i]=le.transform(df[i])
    print(i,df[i].unique())
```

## 3. Modeling Training

### 3.1 Split Data into training and testing set

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('Academic Performance Change', axis=1), df['Academic Performance Change'], test_size=0.2, random_state=0)
```

```python
X_train.shape, X_test.shape
```

### 3.2 Model Building
I will be using the following models to predict the academic performance change for students:
* Random Forest Classifier
* Decision Tree Classifier
* Logistic Regression
* K Nearest Neighbors

#### Random Forest Classifier

```python
rfc = RandomForestClassifier()
```

```python
#Parameter grid
param_grid = {
    'max_depth': [4,8,12,16],
    'min_samples_leaf': [2,4,6,8],
    'min_samples_split': [2,4,6,8],
    'criterion': ['gini', 'entropy'],
    'random_state': [0,42]
}

#GridSearchCV object
grid = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2, scoring='accuracy')

#Fitting the model
grid.fit(X_train, y_train)

#Best parameters
print('Best parameters: ', grid.best_params_)
```

```python
#Random Forest Classifier Object
rfc = RandomForestClassifier(criterion='gini', max_depth=8, min_samples_leaf=8, min_samples_split=2, random_state=42)

#Fitting the model
rfc.fit(X_train, y_train)
```

```python
#Training accuracy
print('Training accuracy: ', rfc.score(X_train, y_train))
```

```python
#predicting the test set results
rfc_pred = rfc.predict(X_test)
```
